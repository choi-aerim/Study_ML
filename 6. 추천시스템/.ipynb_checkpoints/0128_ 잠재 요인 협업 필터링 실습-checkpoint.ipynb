## 경사하강법을 활용한 행렬분해로 잠재요인 찾기

### 경사하강법 평가 방법 사용자함수화

```python
from sklearn.metrics import mean_squared_error
```

```python
def get_rmse(R, P, Q, non_zeros):
    # 예측된 평점 행렬
    pred_R = P.dot(Q.T)
    
    # 평점 행렬 내 0이 아닌 행렬 추출
    non_zero_R = R[R > 0].tolist()
    non_zero_pred_R = [pred_R[i, j] for i, j, r in non_zeros]
    
    mse = mean_squared_error(non_zero_R, non_zero_pred_R)
    rmse = np.sqrt(mse)
    
    return rmse
```
